```python
import os
import sys
import requests
from datetime import datetime
from multiprocessing import Pool
import numpy as np
import pandas as pd
from pathlib import Path
import pickle
from time import time
from typing import Dict, List, Optional
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.pyplot as plt

from tqdm import tqdm
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'
```

```python
S=pd.read_csv("SA0.csv")
img=Image.new("CMYK", (10,50), color=(0,0,0,0))

    

#als.pattern(s[4999:, 0]).name
for i in tqdm(range(51)):
    if(i==0):
        continue
    C=0
    M=0
    Y=0
    K=0
    Const=255
   
        
    cc=0
    mm=0
    yy=0
    kk=0
    C=S["SWS"][i]
    M=S["RESTING"][i]
    Y=S["AWAKE"][i]
    cc=Const*(C/(C+M+Y))
    mm=Const*(M/(C+M+Y))
    yy=Const*(Y/(C+M+Y))
    cc=int(cc)
    mm=int(mm)
    yy=int(yy)
    kk=int(kk)
        
    for z in range(10):
        ZZ=int(z)
        II=int(i-1)
        img.putpixel((ZZ,II),(cc,mm,yy,kk))

                         
plt.imshow(img)
plt.ylabel("g_kvhh")
plt.savefig("g_kvhh.pdf")
        
```

```python
S=pd.read_csv("SA0.csv")
plt.figure(figsize=(10,5))    
T=np.linspace(-2,2,50)
xx=np.linspace(-2,2,10)
X=[]
for i in range(10):
    X.append(xx[i])
s=[]
a=[]
r=[]
e=[]
for i in range(len(S["SWS"])):
    if(i==0):
        continue
    s.append(S["SWS"][i])
    a.append(S["AWAKE"][i])
    r.append(S["RESTING"][i])
    e.append(S["ELSE"][i])
    XXXXX=(s[i-1]+a[i-1]+r[i-1]+e[i-1])
    s[i-1]=s[i-1]/XXXXX
    a[i-1]=a[i-1]/XXXXX
    r[i-1]=r[i-1]/XXXXX
    e[i-1]=e[i-1]/XXXXX
        

fig=plt.figure(figsize=(10,5))   
ax = fig.add_subplot(541)
plt.plot(T, s,  'c',  label="SWS")
plt.plot(T, a, 'y',   label="AWAKE")
plt.plot(T, r, 'm',    label="RESTING")
plt.plot(T,e,'g',  label="ELSE")
plt.xticks(X)
plt.xlabel("Original Value*10^")
plt.ylabel("Numbers")
plt.title("g_leak")
Z=np.linspace(-2,2,5)
ax.set_xticks(Z)
Z=np.linspace(0,1,3)
ax.set_yticks(Z)
plt.savefig("g_leak.pdf")
```
