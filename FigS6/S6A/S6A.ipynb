```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
plt.rcParams['font.family'] = 'Arial'

T2 = pd.read_csv("A2.csv")
T3 = pd.read_csv("A3.csv")
T4 = pd.read_csv("A4.csv")
T5 = pd.read_csv("A5.csv")
T6 = pd.read_csv("A6.csv")
T7 = pd.read_csv("A7.csv")

s2 = []
a2 = []
r2 = []
e2 = []

s3 = []
a3 = []
r3 = []
e3 = []

s4 = []
a4 = []
r4 = []
e4 = []

s5 = []
a5 = []
r5 = []
e5 = []

s6 = []
a6 = []
r6 = []
e6 = []

s7 = []
a7 = []
r7 = []
e7 = []

for i in range(101):
    if(i==0):
        continue
    s2.append(T2["SWS"][i])
    a2.append(T2["AWAKE"][i])
    r2.append(T2["REST"][i])
    e2.append(T2["EXCLUDED"][i])
    s3.append(T3["SWS"][i])
    a3.append(T3["AWAKE"][i])
    r3.append(T3["REST"][i])
    e3.append(T3["EXCLUDED"][i])
    s4.append(T4["SWS"][i])
    a4.append(T4["AWAKE"][i])
    r4.append(T4["REST"][i])
    e4.append(T4["EXCLUDED"][i])
    s5.append(T5["SWS"][i])
    a5.append(T5["AWAKE"][i])
    r5.append(T5["REST"][i])
    e5.append(T5["EXCLUDED"][i])
    s6.append(T6["SWS"][i])
    a6.append(T6["AWAKE"][i])
    r6.append(T6["REST"][i])
    e6.append(T6["EXCLUDED"][i])
    s7.append(T7["SWS"][i])
    a7.append(T7["AWAKE"][i])
    r7.append(T7["REST"][i])
    e7.append(T7["EXCLUDED"][i])
```

```python
List=[]
for i in range(100):
    List.append([s2[i],a2[i],r2[i],e2[i]])
List.sort()
    
data = {}
for i in range(100): 
    data[str(i)] =List[i]

df = pd.DataFrame(data)

row1_percent = df.iloc[0] 
row2_percent = df.iloc[1] 
row3_percent = df.iloc[2] 
row4_percent = df.iloc[3] 


plt.figure(figsize=(10, 15))

index = np.arange(len(df.columns))
bar_width = 0.35


plt.barh(index, row1_percent, bar_width, label='SWS')
plt.barh(index, row2_percent, bar_width, left=row1_percent, label='AWAKE')
plt.barh(index, row3_percent, bar_width, left=row1_percent+row2_percent, label='RESTING')
plt.barh(index, row4_percent, bar_width, left=row1_percent+row2_percent+row3_percent, label='EXCLUDED')

plt.ylabel('Columns')
plt.xlabel('Percentage')
plt.title('Normal')
Z=np.linspace(0,84,3)
plt.xticks(Z)
#plt.yticks(index, df.columns)
plt.legend()

plt.tight_layout()

plt.savefig("Fig6C1.pdf")
```
