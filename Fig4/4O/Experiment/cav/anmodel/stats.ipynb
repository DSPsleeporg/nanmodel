```python
import pandas as pd
import matplotlib.pyplot as plt
N=5
S=pd.read_csv("OD{}.csv" .format(str(N)))
A=pd.read_csv("OD1{}.csv" .format(str(N)))
A.head(5)
```

```python
ms=[]
ma=[]
ss=[]
sa=[]
for i in range(len(S["mean"])):
    if(S["mean"][i]!=0 and A["mean"][i]!=0 and S["mean"][i]!=-1 and A["mean"][i]!=-1):
        ms.append(S["mean"][i])
        ma.append(A["mean"][i])
        ss.append(S["stdev"][i])
        sa.append(A["stdev"][i])
plt.scatter(ms,ma)
plt.xlabel("sws_median ISI")
plt.ylabel("awake_median ISI")
```

```python
import numpy as np
print(np.mean(ms))
print(np.mean(ma))
print(np.std(ms))
print(np.std(ma))
```
