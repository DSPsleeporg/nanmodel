```python
N=0
A=[0,0,0]
BBB=['mean', 'median', 'stdev']
dic={key:  val for key, val in zip(BBB,A)}
new_params: pd.DataFrame=pd.DataFrame.from_dict(dic,orient='index').T
new_params.to_csv("OD{}.csv" .format(str(N)))
```

```python
x=len(S["g_kvhh"])
DD=pd.read_csv("NAN{}.csv" .format(N))
for j in tqdm(range(x)):
    if j==0:
        continue   
    if(DD["ans"][j]==0):
        A=[-1,-1,-1]
        BBB=['mean', 'median', 'stdev']
        dic={key:  val for key, val in zip(BBB,A)}
        new_params=dic
        T=pd.read_csv("OD{}.csv" .format(str(N)))
        Nn: pd.DataFrame=pd.DataFrame.from_dict(new_params, orient="index").T
        T=pd.concat([T,Nn])
        T.to_csv("OD{}.csv" .format(str(N)))
        continue 
        
    A=[]
    c=j
    A.append(S['g_kvhh'][c])
    A.append(S['g_unav'][c])
    A.append(S['g_kna'][c])
    A.append(S['g_leak'][c]*0.6095)
    A.append(S['g_leak'][c]*0.3905)
    A.append(S['g_cav'][c])
    A.append(S['x_na'][c])
    A.append(S['y_na'][c])
    A.append(S['t_na'][c])
    BBB=['g_kvhh', 'g_unav', 'g_kna', 'g_lek','g_lena', 'g_cav', 'x_na', 'y_na', 't_na']
    
    A[N]=A[N]*1.0
    dic={key:  val for key, val in zip(BBB,A)}
    #new_params: pd.DataFrame=pd.DataFrame.from_dict(dic,orient='index').T
    new_params=dic

    mnnan1.set_params(new_params)
    s,info=mnnan1.run_odeint()
    X=s[9999:,3]
        
    CCC=[]
    ss=signal.argrelmax(s[9999:,0])
    for k in range(len(ss[0])-1):
        if(ss[0][k+1]-ss[0][k]>80):
            continue
        CCC.append(ss[0][k+1]-ss[0][k])
    if(len(CCC)<3):
        A=[0,0,0]
        BBB=['mean', 'median', 'stdev']
        dic={key:  val for key, val in zip(BBB,A)}
        new_params=dic
        T=pd.read_csv("OD{}.csv" .format(str(N)))
        Nn: pd.DataFrame=pd.DataFrame.from_dict(new_params, orient="index").T
        T=pd.concat([T,Nn])
        T.to_csv("OD{}.csv" .format(str(N)))
        continue 
            
    A=[statistics.mean(CCC),statistics.median(CCC),statistics.stdev(CCC)]
    BBB=['mean', 'median', 'stdev']
    dic={key:  val for key, val in zip(BBB,A)}
    new_params=dic
    T=pd.read_csv("OD{}.csv" .format(str(N)))
    Nn: pd.DataFrame=pd.DataFrame.from_dict(new_params, orient="index").T
    T=pd.concat([T,Nn])
    T.to_csv("OD{}.csv" .format(str(N)))
    continue    
        
    
        
```
